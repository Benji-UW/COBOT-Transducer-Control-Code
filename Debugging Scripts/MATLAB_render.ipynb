```python
import numpy as np
import scipy
from scipy.io import loadmat
import matplotlib.pyplot as plt
import os

plt.style.use("default")
# plt.rcParams['figure.dpi'] = 250

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
    'figure.dpi': 250})
```

```python
path = os.path.abspath('..\\')
mat = loadmat(path + "\\Scans\\scan_2x.mat")
mdat = mat['chA']
```

```python
time_axis = np.linspace(0,1e2,mdat.size) - 5
size = mdat.size

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()

ax.plot(time_axis, mdat, linewidth=0.2)
ax.vlines(5,-600,600,ls='--',color='gray', linewidth=0.5,label='cutoff threshold')
ax.hlines(47,-100,100,color='r', linewidth=0.5)
ax.scatter(time_axis[np.argmax(mdat[size//10:]) + size//10], np.max(mdat[size//10:]),marker='*',color='C1',label='Signal max')

ax.set_ylim(-600,600)
ax.set_xlim(time_axis[0],time_axis[-1])
ax.set_ylabel("Signal magnitude (mV)")
ax.set_xlabel("Time $(\mu s)$")
ax.legend()
ax.set_title("Sample output from immersion transducer")

# fig.savefig("figures\\IT_sample_1.jpg", dpi=200)
```
